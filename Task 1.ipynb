```python
#Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading Data

data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
print("The given dataset is \n")
print(data)
print("\n")

print(data.describe())
```

```python
#Plotting the Dataset

from sklearn.linear_model import LinearRegression
lm = LinearRegression()

X = data[['Hours']]
Y= data['Scores']

lm.fit(X,Y)
print("\n")
print("Intercept",lm.intercept_)
print("Coefficient of x or slope",lm.coef_)

print("\n")

data.plot(x='Hours',y='Scores',style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours')
plt.ylabel('Percentage')
plt.show()
```

```python
#Checking the accuracy of our algorithm

pred = lm.predict(X)
#df=pd.data({'Actual':Y, 'Predicted':pred})
df = pd.DataFrame({ 'Actual': Y, 'Predicted': pred})
print(df)
df.plot(kind="bar")
print("\n")

from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(Y, pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y, pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y, pred)))
```

```python
#Predicting the percentage when hours studied = 9.5

hrs=np.array([9.25])
hrs=hrs.reshape(-1,1)
predicted=lm.predict(hrs)
print("Predicted percentage if a student studies for 9.25 hours in a day",predicted)
```
